```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

# Loading the Data 

```python
df=pd.read_csv('tmdb_movies_data.csv', sep=',')
df 
```

```python
df.head()
```

# About the dataset

1. imdb_id: It represents imdb_id of the movie in the IMDB Database.
2. popularity: It represents the popularity score of the movie.

3. budget: It represents the budget of the movie in dollars.

4. revenue: It represents the total revenue collected by the movie in dollars.

5. original title: It represents the title of the movie.

6. cast: It tells the actors who acted in that movie.

7. director: It tells the director of the movie.

8. overview: It tells the overview of the plot of the movie.

9. runtime: It tells the overall runtime of the movie in minutes.

10. genres: It represents the genres the movie represents.

11. release date: It tells the date of release of the movie.

12. release year. It tells the year in which the movie was released.

# Data Wrangling 

```python
df.head(20)
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True) #droping duplicates
df.info()
```

```python
md=df[['popularity','budget','revenue','original_title','runtime','genres','vote_count','vote_average','release_year']]
```

```python
md['profit']=md['revenue']-md['budget']
```

```python
md.info()
```

```python
md.describe()
```

```python
md.describe(include ="all")
```

```python
md.isnull().sum()
```

```python
 md.dropna(subset=['genres'],inplace=True)
```

```python
md.info()
```

```python
md.isnull().sum()
```

```python
md.head()
```

```python
#split genre and create a new entry for each of the genre a movie falls into 
s=md['genres'].str.split('|').apply(pd.Series,1).stack()
print(s)
s.index=s.index.droplevel(-1)
s.name='genres'
del md['genres']
md=md.join(s)
```

```python
md.head()
```

```python
md.shape
```

```python
corr=md.drop(['original_title','genres'],axis=1,inplace=False).corr()
sns.heatmap(corr,annot=True)#we can use cmap for further color option 
```

## question 1. Which Genres are the most common (number of movie made)?


```python
num_movie_genres=md_genres.original_title.nunique().sort_values(ascending=False)
print(num_movie_genres)
```

```python
plt.figure(figsize=(8,8))
plt.pie(x=num_movie_genres.values,labels=num_movie_genres.index,autopct="%1.1f%%")
plt.title("Movie per genre in %")
plt.show()
```

# Which genres have high avg. Budget and Revenue?

```python
sorted_by_budget=md_genre_budget.sort_values(by='budget',ascending=False)

md_genre_revenue=pd.DataFrame(md_genres['revenue'].mean()).reset_index()
sorted_by_revenue=md_genre_revenue.sort_values(by='revenue',ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x='genres',y='budget',data=sorted_by_budget)
plt.title('Average Movie Budget by genres (hidhest First)')
plt.xlabel("genres")
plt.ylabel('Average Budget')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


plt.figure(figsize=(10,5))
sns.barplot(x='genres',y='revenue',data=sorted_by_revenue)
plt.title('Average Movie revenue by genres (highest First)')
plt.xlabel("genres")
plt.ylabel('Average revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# in this graph above we can clearly see that the genre adventure has both the highest avg. budget and revenue , fantasy comes second in budget and revenue. 



































# Q3. Which Genre have high avg. profit?

# Q4. which genres have high number of movies with voting avg. >=8?
